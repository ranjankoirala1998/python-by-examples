<h3>Challenge 140</h3>
<p>Using the PhoneBook database from program 139, write a program that will display the following menu.</br></br>Main Menu</br></br>1) View Phone book</br>2) Add to phone book</br>3) Search for surname</br>4) Delete person from book</br>5) Qit</br></br>Enter your selection : </p>
 
<p>If the user selects 1, they should be able to view the entire phonebook. If they select 2, it should allow them to add a new person to the phonebook. If they select 3, it should ask them for a surname and then display only the records of people with the same surname. If they select 4, it should ask for an ID and then delete that record from the table. If they select 5, it should end the program. Finally, it should display a suitable message if they enter an incorrect selection from the menu. They should return to the menu after each action, until they select 5.</p>

```python
import sqlite3

with sqlite3.connect(database = 'files/phonebook.db') as db:
    cursor=db.cursor() 

def view_phone() : 
    cursor.execute("SELECT * FROM Names;")
    print("""
------------------------------------------
 ID | firstname | lastname | phone_number      
------------------------------------------
        """)
    for name in cursor.fetchall() :
        print(name)

def add_to_phone() : 
    name_id = int(input("Enter ID : "))
    firstname = input("Enter Name : ")
    lastname = input("Enter Surname : ")
    phone_number = input("Enter Number : ")

    cursor.execute("""
    INSERT INTO Names(id, first_name, surname, phone_number)
    VALUES(?, ?, ?, ?)""", (name_id, firstname, lastname, phone_number))
    db.commit()

def search_surname() :
    surname = input("Enter Surname : ")

    cursor.execute("SELECT * FROM Names WHERE surname = ?;", [surname])

    print("""
------------------------------------------
 ID | firstname | lastname | phone_number      
------------------------------------------
        """)
    for name in cursor.fetchall() :
        print(name)

def delete_person() : 
    name_id = int(input("Enter ID : "))

    cursor.execute("DELETE FROM Names WHERE id = ? ;", [name_id])

    db.commit()

def quit_now() : 
    return False

flag = True

while flag : 
    print("""
    Main Menu

    1) View Phone book
    2) Add to phone book
    3) Search for surname
    4) Delete person from book
    5) Quit\n
    """)

    choice = int(input("Enter your selection : "))

    if choice == 1 :
        view_phone()

    elif choice == 2 :
        add_to_phone()
    
    elif choice == 3 :
        search_surname()
    
    elif choice == 4 :
        delete_person()

    elif choice == 5 :
        flag = quit_now()

    else : 
        print("Invalid Option!!")

db.close()
```

<h3>Challenge 142</h3>
<p>Using the BookInfo database from program 141, display the list of authors and their place of birth. Ask the user to enter a place of birth and then show the title, date published and author’s name for all the books by authors who were born in the location they selected.</p>

```python
import sqlite3

with sqlite3.connect("files/BookInfo.db") as db :
    cursor = db.cursor()

cursor.execute("SELECT * FROM Authors")
print("""
-------------------------------
   name   |   place_of_birth   
-------------------------------""")
for author in cursor.fetchall() : 
    print(author)

place = input("Enter the place of birth : ")

print("\nBooks by authors born in {}.".format(place))

print("""
---------------------------------------------
   title   |   author   |   date_published   
---------------------------------------------""")

cursor.execute("""
SELECT B.title, A.name, B.date_published
FROM Authors A
INNER JOIN Books B
ON A.name = B.author
WHERE A.place_of_birth == ?""", [place])

for record in cursor.fetchall() :
    print(record)

db.close()
```

<h3>Challenge 143</h3>
<p>Using the BookInfo database, ask the user to enter a year and display all the books published after that year, sorted by the year they were published.</p>

```python
import sqlite3

with sqlite3.connect("files/BookInfo.db") as db :
    cursor = db.cursor()


year = input("Enter an year : ")

print("All the books published after the year {} are : ".format(year))

cursor.execute("""
SELECT title, date_published
FROM Books
WHERE date_published > ? OR date_published == ?
ORDER BY date_published
""", [year, year])

print("""
--------------------------------
   title   |   date_published      
--------------------------------""")

for record in cursor.fetchall() :
    print(record)

db.close()
```

<h3>Challenge 144</h3>
<p>Using the BookInfo database, ask the user for an author’s name and then save all the books by that author to a text file, with each field separated by dashes so it looks as follows:</p>
<img src="files/c144.png"/>
<p>Open the text file to make sure it has worked correctly.</p>

```python
import sqlite3

with sqlite3.connect("files/BookInfo.db") as db :
    cursor = db.cursor()

file = open("files/BookByAuthor.txt", "w")


author = input("Enter an author's name : ")

cursor.execute("""
SELECT *
FROM Books
WHERE author = ?
""", [author])

for record in cursor.fetchall() :
    file.write(str(record[0])+' - '+record[1]+' - '+record[2]+' - '+record[3]+'\n')

file.close()
db.close()
```
